Введем новую переменную "PrizeCategory", которая отражает, является ли полученная премия результатом работы одного ученого или команды.

```python
def get_prize_cat(sh):
    if sh < 2:
        return 'Individual'
    elif sh >=2:
        return "Collective"
    else:
        return sh
```

```python
df_nobel ['PrizeCategory'] = df_nobel['share'].apply(get_prize_cat)
```

```python
df_nobel.head()
```

```python
df_nobel ['PrizeCategory'] = pd.Categorical(df_nobel ['PrizeCategory'])
```

# Z-оценка в pandas (Способ 1)

Произвести z-оценку в pandas можно несколькими способами. На примере первого дата-сета в этой работе показан Способ 1. Для демоснтрации владения спсобом 2 был выбран другой дата сет с более релевантными данными.

```python
import pandas as pd
import numpy as np
import scipy.stats
```

```python
nobel = pd.read_csv('nobel_prize_by_winner.csv')
nobel
```

```python
X = nobel ['share'] [0]
mean = nobel ['share'].mean()
std = nobel ['share'].std(ddof=0)
print (X, mean, std)
```

```python
z_score = (X - mean) / std
print (z_score)
```

# Z-оценка в Pandas (Способ 2)

Для демонстрации владения навыком проведения z-оценки в pandas необходимо использовать еще один дата сет с более релевантными данными. В дата сете Exams оценены навыки студентов в трех различных направлениях: вычисление, письмо и чтение.

Исследуем гипотезу, что девочки имеют более высокие результаты в вычислениях и письме. Для этого, произведем z-оценку их результатов, и сравним средние значения. 

```python
import pandas as pd
import numpy as np
import scipy.stats
```

```python
exam = pd.read_csv('exams.csv')
exam
```

```python
exam['math score'].hist()
```

```python
exam['writing score'].hist()
```

Для удобства создадим новые дата фреймы для результатов по вычислениям и письму.

```python
math = exam[["math score", "gender"]]
```

```python
math.head()
```

```python
writing = exam[["writing score", "gender"]]
```

```python
writing.head()
```

Произведем z-оценку данных.

```python
scipy.stats.zscore(math['math score'])
```

```python
math ['z-score'] = scipy.stats.zscore(math['math score'])
```

```python
math.head()
```

```python
scipy.stats.zscore(writing['writing score'])
```

```python
writing ['z-score'] = scipy.stats.zscore(writing['writing score'])
```

```python
writing.head()
```

```python
exam_new = pd.concat([math, writing])
exam_new
```

```python
Изучим выбросы наших данных.
```

```python
exam_new[exam_new['z-score'] < -3]
```

Проверим заявленную гипотезу.

```python
exam_new.groupby('gender')['z-score'].mean()
```

Согласно проведенной z-оценке результаты девочек по исследуемым дисциплинам дейтсвиетльно выше, чем у мальчиков.

# Разведочный анализ данных

Для того, чтобы иметь развернутое представление об исследуемых данных, можно использовать различные инструменты их визуализации. Более того, методы разведочного анализа данных часто позволяют подтвердить или опровергнуть различные гипотезы и найти интересные взаимосвязи.
В данном случае, попробуем проверить следующую гипотезу: студенты, которые прошли подготовительный курс, имеют более высокие результаты за вычмслительный блок заданий.

```python
import seaborn as sns
```

Для начала, определим в выбранном дата сете категориальные и числовые переменные.

```python
num_features = ['math score', 'reading score', 'writing score']
cat_features = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course' ]
```

```python
Далее, посмотрим на распределение наблюдение результатов по вычислениям,
```

```python
sns.distplot(exam['math score'])
```

Оценим, сколько студентов прошли подготовительный тест.

```python
exam['test preparation course'].value_counts()
```

```python
sns.barplot(x=exam['test preparation course'].value_counts().index, y=exam['test preparation course'].value_counts())
```

Наконец, построим скрипичный график для оценки взаимосвязи факта прохождения студентом подготовительного теста и его результатом за вычислительный блок.

```python
sns.violinplot(x=exam['test preparation course'], y=exam['math score'])
```

Данный не дает возможности до конца подтвердить или опровергнуть выдвинутую гипотезу, однако подсвечивает интересный вывод, что студенты, которые прошли подготовительный курс, часто достигали результата от 70 до 80 баллов за вычислительный блок.

Далее, попробуем оценить коллеряцию результатов студентов в различных блоках между собой, отталкиваясь от гипотезы, что все они находятся в прямой зависимости: чем выше оценка в одном блоке, тем вероятнее высокий результат в двух других.

```python
correlation = exam.corr().loc[num_features, num_features]
correlation
```

```python
fix, ax = plt.subplots(figsize = (14,12))

sns.heatmap(correlation, annot=True, vmax=1, vmin=-1)
```

В результате, мы получили очень высокие корреляции между исследуемыми переменными, что подтверждает нашу гипотезу. Чем выше у студента оценка за отдельный блок, тем выше вероятность его высокого результата в двух других.

Чтобы продолжить исследование данных по результатам вычислительного блока экзамена, попробуем построить взаимосвязи этих наблюдений с остальными данными о студенте и сделать ряд соотвествующих выводов.

```python
df_math = pd.melt(exam, id_vars='math score', value_vars=cat_features)
```

```python
def boxplot(x,y, **kwargs):
    sns.boxplot (x=x, y=y)
    x=plt.xticks(rotation=90)
    
g = sns.FacetGrid (df_math, col='variable',
                col_wrap=2, sharex=False, sharey=False, height=5)

g.map(boxplot, "value", 'math score')
        
```

Построение подобной системы графиков позволяет отметить интересные результаты. Результаты молодых людей за вычислительный блок выше, чем результаты девушек.
Также, результаты тех студентов, кто проходил подготовительный курс, также оказались несколько выше. Самые высокие результаты среди всех студентов показали те, кто имеет степени бакалавра или магистра. 
Отдельного внимания заслуживает необычное наблюдение о том, что те, кто мог позволить себе платный ланч, продемонстрировали более высокий балл за выполнение вычислительного блока теста!
